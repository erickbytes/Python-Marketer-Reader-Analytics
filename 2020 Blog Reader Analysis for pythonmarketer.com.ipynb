```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
url = 'https://raw.githubusercontent.com/erickbytes/Python-Marketer-Reader-Analytics/master/dataset/pythonmarketer.com_countryviews_year_01_01_2020_12_31_2020.csv'
readers = pd.read_csv(url, names=['country','views'])
readers.head()
```

```python
# top 25 views by country in 2020 on pythonmarketer.com
readers.nlargest(25, 'views')
```

```python
url = 'https://github.com/erickbytes/Python-Marketer-Reader-Analytics/blob/master/dataset/2020_pythonmarketer.com_post_views.xlsx?raw=true.xlsx'
posts = pd.read_excel(url)
posts.head()
```

```python
# extract date each post was published from post urls
posts['date'] = posts.url.str.split(pat='/', n=6).apply(lambda url:[x for x in url if x.isdigit()])
posts.date = pd.to_datetime(posts.date.str.join('-'))
```

```python
posts.head()
```

```python
# top 25 most viewed posts in 2020 on pythonmarketer.com
posts.nlargest(25, 'views')
```

```python
# posts read in 2020 by year published
posts.date.dt.year.fillna(0).astype(int).value_counts()
```
