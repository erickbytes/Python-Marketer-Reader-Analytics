## Most Read Posts on 2020 on pythonmarketer.com

```python
url = 'https://github.com/erickbytes/Python-Marketer-Reader-Analytics/blob/master/dataset/2020_pythonmarketer.com_post_views.xlsx?raw=true.xlsx'
posts = pd.read_excel(url)
posts.head()
```

```python
# extract date each post was published from post urls
posts['date'] = posts.url.str.split(pat='/', n=6).apply(lambda url:[x for x in url if x.isdigit()])
posts.date = pd.to_datetime(posts.date.str.join('-'))
posts.info()
```

```python
# Top 25 most viewed posts in 2020 on pythonmarketer.com
posts.nlargest(25, 'views')
```

```python
# posts by year published, since I began writing in 2016
ax = (posts.date.dt.year.fillna(0).astype(int)
                                  .value_counts(sort=False)
                                  .drop([0])
                                  .plot(kind='bar', rot=45, title='pythonmarketer.com Posts By Year Published'))
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```

## Historically, spring has been the most productive season for  pythonmarketer.com 

```python
# Historically, August, May, April and March are tied for most posts.
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July','August','September','October','November','December']
posts_by_month = (posts.date.dt.month
                        .fillna(0)
                        .astype(int)
                        .value_counts(sort=False))
      
ax = posts_by_month.plot(kind='bar',
                         rot=45,
                         xticks=posts_by_month.index,
                         title='Posts Published by Month on pythonmarketer.com - 2016-2020')
ax.set_xticklabels(months)
ax
```

```python
# top 25 most viewed posts by year published
(posts.nlargest(25, 'views').date.dt.year.fillna(0)
                                         .astype(int)
                                         .value_counts(sort=False)
                                         .drop([0])
                                         .plot(kind='bar', rot=45, title='pythonmarketer.com Most Read Posts, Year Published'))
```

```python
# plot 2020 page views based on the date the post viewed was published
posts = posts.sort_values(['date'])
#plt.set_title('subplot 2')
plt.plot_date(x=posts.date, y=posts.views, markersize=10)
plt.title('Posts Viewed in 2020 Vs. Date Published - pythonmarketer.com')
```

# pythonmarketer.com View Referral Sources
Most readers find pythonmarketer.com through search engines like duckduckgo.com, Google, Bing, Yahoo! and Baidu to name a few of them. qwant.com, for example (15 views), is a French search engine.

```python
url = 'https://raw.githubusercontent.com/erickbytes/Python-Marketer-Reader-Analytics/master/dataset/pythonmarketer.com_referrers_year_01_01_2020_12_31_2020.csv'
referrers = pd.read_csv(url, names=['source','views'])
referrers.sort_values('views', ascending=False).reset_index(drop=True).head(50)
```

```python
referrers.nlargest(10, 'views').plot(kind='bar')
```
