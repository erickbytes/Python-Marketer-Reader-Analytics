```python
import pandas as pd
```

```python
readers = pd.read_csv("countries.csv")
#readers.style.format('{:,}')
pd.options.display.float_format = '{:,}'.format
readers.head()
```

```python
readers = readers.sort_values(by="Fert. Rate", ascending=False)
readers.head(20)
```

```python
def replace_na(age):
    if 'N.A.' in str(age):
        return float(0)
    else:
        return age
    
readers['Med. Age'] = readers['Med. Age'].apply(replace_na)
readers['Med. Age'] = readers['Med. Age'].fillna(0).astype(float, errors='raise')
readers.nlargest(columns='Med. Age', n=15).reset_index(drop=True)
```

```python
readers['LearnerRatioPerCapitaInverse'] =  readers["visitors"] / readers["Population (2020)"]
#readers["Population (2020)"] = readers["Population (2020)"]('{:,}')
PopulationNovelty = readers[['country','visitors','LearnerRatioPerCapitaInverse','Population (2020)']].sort_values(by='LearnerRatioPerCapitaInverse', ascending=False)
```

```python
PopulationNovelty.nlargest(columns='LearnerRatioPerCapitaInverse', n=25)
```

```python
PopulationNovelty["LearnerRatioPerCapitaInverse"] = PopulationNovelty["LearnerRatioPerCapitaInverse"].apply(lambda x:x*100000000)
PopulationNovelty["LearnerRatioPerCapitaInverse"] = PopulationNovelty["LearnerRatioPerCapitaInverse"].fillna(0).astype(int)
PopulationNovelty.nsmallest(columns='LearnerRatioPerCapitaInverse', n=25)
```
